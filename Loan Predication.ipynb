```python
data.dropna(inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import LabelEncoder

cols=['Gender','Married','Education','Self_Employed','Loan_Status']

for col in cols:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])


Ordinal_Encoder=OrdinalEncoder(categories=[['Rural','Semiurban','Urban']])
data['Property_Area']=Ordinal_Encoder.fit_transform(data[['Property_Area']])


data=data.replace(to_replace='3+',value=4)

data=data.drop(columns=['Loan_ID'],axis=1)
```

```python
Y=data['Loan_Status']
new_data=data.drop('Loan_Status',axis=1)

new_data.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()

scaled_data=scaler.fit_transform(new_data)
scaled_data
```

```python
data_df=pd.DataFrame(scaled_data,columns=new_data.columns)
data_df.head()
```

```python
X=data_df
```

```python
X
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
classifier=sm.SVC(kernel='linear')
classifier.fit(x_train,y_train)
y_pred = classifier.predict(x_test)
x_pred = classifier.predict(x_train)
```

### Training Accuracy

```python
Training_Accuracy=accuracy_score(x_pred,y_train)
Training_Accuracy
```

### Testing Accuracy

```python

Testing_Accuracy=accuracy_score(y_pred,y_test)
Testing_Accuracy
```

```python
cm=confusion_matrix(y_pred,y_test)
print("Confusion Matrix:\n", cm)
    
```
